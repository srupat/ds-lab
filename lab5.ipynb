```r
df <- read.csv('knn1_csv.csv')
df
```

```r
x <- 3
y <- 2
dist <- sqrt((x - df$x)**2 + (y - df$y)**2)
df <- cbind(df, dist)
df
```

```r
df <- df[order(df$dist),]
df
```

```r
# NN
cat("\nclass of P for NN is:", df[1,4])

# KNN k = 5
df1 <- df[1:5,]
l1 <- sum(df1$class == 1)
l2 <- sum(df1$class == 2)
l3 <- sum(df1$class == 3)

if(l3>l1 & l3>l2){
	cat("\nclass of P for KNN(k=5) is:", l3)
}
if(l2>l1 & l2>l3){
	cat("\nclass of P for KNN(k=5) is:", l2)
}
if(l1>l2 & l1>l3){
	cat("\nclass of P for KNN(k=5) is:", l1)
}


# KNN k = 7
df2 <- df[1:7,]
l11 <- sum(df2$class == 1)
l22<- sum(df2$class == 2)
l33 <- sum(df2$class == 3)

if(l33>l11 & l33>l22){
	cat("\nclass of P for KNN(k=7) is:", l3)
}
if(l22>l11 & l22>l33){
	cat("\nclass of P for KNN(k=7) is:", l2)
}
if(l11>l22 & l11>l33){
	cat("\nclass of P for KNN(k=7) is:", l1)
}

# RNN (r = 1.45)
df3 <- df[df$dist < 1.45, ]
l111 <- sum(df2$class == 1)
l222 <- sum(df2$class == 2)
l333 <- sum(df2$class == 3)

if(l333>l111 & l333>l222){
	cat("\nclass of P for RNN is:", l3)
}
if(l222>l111 & l222>l333){
	cat("\nclass of P for RNN is:", l2)
}
if(l111>l222 & l111>l333){
	cat("\nclass of P for RNN is:", l1)
}

# MKNN (k = 5)
dk <- max(df$dist[1:5])

d_1 <- df$dist[1]

weight <- (dk - df$dist) / (dk - d_1)

df <- cbind(df, weight)

new_df <- head(df, 5)

wc1 <- sum(new_df$weight[new_df$class == 1])
wc2 <- sum(new_df$weight[new_df$class == 2])
wc3 <- sum(new_df$weight[new_df$class == 3])

cat("\nTotal weight for class 1:", wc1, "\nTotal weight for class 2:", wc2, "\nTotal weight for class 3:", wc3)

if (wc3 > wc2 & wc3 > wc1) {
	cat("\nTherefore class of P for MKNN(k=5) is: 3")
} else if (wc2 > wc1 & wc2 > wc3) {
	cat("\nTherefore class of P for MKNN(k=5) is: 2")
} else {
	cat("\nTherefore class of P for MKNN(k=5) is: 1")
}
```
